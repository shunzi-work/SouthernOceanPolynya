```python
# Calculate the autocorrelation at lags from 0.01 (one time-step) to 6 (about a third of the series)
# Create a sine function evaluated every 0.01 from 0 to 100
step = 1
t2 = np.arange(0,500,step)
y1 = ds_polynya_ann

lag_min = 5
lag_max = 200
auto1 = []
lags1 = []
for lag in range(lag_min,int(lag_max/step)):
    v = np.corrcoef(y1[:-lag],y1[lag:])
    lags1.append(lag*step)
    auto1.append(v[0,1])

plt.plot(lags1,auto1,'b')
plt.title('autocorrelation of polynya')
plt.xlabel('lag')
plt.ylabel('autocorrelation')
```

```python
X = np.fft.fft(ds_conv_ann)
N = len(X)
n = np.arange(N)
# get the sampling rate
sr = 1 
T = N/sr
freq = n/T 

# Get the one-sided specturm
n_oneside = N//2
# get the one side frequency
f_oneside = freq[:n_oneside]

# convert frequency to hour
t_h = 1/f_oneside 

plt.figure(figsize=(15,4))
plt.plot(t_h, np.abs(X[:n_oneside])/n_oneside)
plt.xticks([12, 24, 84, 168])
plt.xlim(0, 200)
plt.xlabel('Period')
plt.title('convection')
```

```python
fs = 1
s = ds_polynya_ann
ts = np.arange(0, 500, 1/fs)
freqs = np.fft.rfftfreq(len(s), 1/fs)
fft = np.fft.rfft(s)

plt.figure(figsize=(12, 8))
plt.subplot(2,1,1)
plt.plot(ts, s)
plt.title('Time-Domain')
plt.xlabel('Time (sec)')
plt.subplot(2,1,2)
plt.plot(freqs, np.abs(fft))
plt.title('Frequency-Domain')
plt.xlabel('Frequency (Hz)')
plt.tight_layout()
```
